### Limpeza de dados com Pandas

#### 1) Leitura do CSV

```
import pandas as pd

df = pd.read_csv('../data/dados.csv', sep=';', engine='python')
df.head()
```

#### 2) Informações gerais e visualização

```
df.info(memory_usage='deep')
```

#### 3) Subconjunto de colunas

```
sub = df[['ID','Date','Calories']].copy()
sub.head()
```

#### 4) Limpeza: Calories → 0; Date nulos → '1900/01/01'

```
df_limpo = df.copy()
df_limpo['Calories'] = df_limpo['Calories'].fillna(0)
df_limpo['Date'] = df_limpo['Date'].fillna('1900/01/01')
```

#### 5) Corrigir data incorreta '20201226' e converter para datetime

```
df_limpo['Date'] = df_limpo['Date'].replace('20201226', '2020/12/26')
df_limpo['Date'] = pd.to_datetime(df_limpo['Date'], format='%Y/%m/%d', errors='coerce')
df_limpo.info()
```

#### 6) Remover linhas com Date inválida (NaT)

```
df_limpo = df_limpo.dropna(subset=['Date']).reset_index(drop=True)
df_limpo.info()
```

#### 7) Salvar resultado

```
df_limpo.to_csv('../data/dados_limpos.csv', sep=';', index=False)
print('OK, salvo em ../data/dados_limpos.csv')
```
